### List
* Used to store multiple items in single variables
* We can store multiple types of data in list
* List stores data in ordered form, It allows duplicate data
* Elements are separeted by commas, and stored inside square brackets []

```python
list1 = [int]
list2 = []
s = 'Length is: '

#Add new elements or Update list
list1.extend([1,2,3,'a'])
list2.append([1,'a',"abc",1.0])
list2.append(2)

print(type(list1), s, len(list1))
print(type(list2), s, len(list2))
print(list1)
print(list2)

#Overwrite the List
list1 = [0,1,1,1,1]

#Delete list
del list1[0]
print(list1)
```

### Tuple
* This is same as list but immutable
* We can't modify data once we difined it.
* same as List it stires data in ordered form and allows duplicate data
* To define a tuple comma is important, Comma defines tuple, without comma it considered as other data type. ex..., Tup = (1,)
* It stores data inside round brackets()

```python
tup = tuple()
tup = (0, 'a', 1.2,0)
tup = (1,'b',2.2,1)
print(tup.count(0))
print(type(tup))
print(tup)
print(tup[1])
```

### Dictionary

* It stores data in Key-Value pairs
* Dictionary data is unordered
* Keys are Unique, But Values can be duplicates
* It is Mutable and Defined using curly braces {}

```python
dictionary = {}
print(type(dictionary))

dictionary = {"Key1" : (1,1,1,1)}
print(dictionary)

#Overwrite the Dictionary
dictionary = {"Key2" : "Vivek Avhad"}
print(dictionary)

#Update the key
dictionary["Key2"] = [1,1,1,1]
print(dictionary)

#Adding new Key
dictionary["Key1"] = [2,2,2,2]
print(dictionary)

#Delete key
del dictionary["Key2"]
print(dictionary)
```
