## Time Complexity
`Measures how the number of basic operations of an algorithm grows as the input size n increases`

## Space Complexity
`Measures how the amount of memory used by an algorithm grows as the input size n increases`

### Big-O Complexity Ranking (Best to Worst)

* **O(1)** - Constant (Best)
* **O(log n)** - Logarithmic
* **O(√n)** - Square Root
* **O(n)** - Linear
* **O(n log n)** - Linearithmic
* **O(n²)** - Quadratic
* **O(n³)** - Cubic
* **O(2ⁿ)** - Exponential
* **O(n!)** - Factorial (Worst)

### 1. O(1) — Constant Time (Best)
`The time stays almost the same no matter how large the input becomes.`

```python
arr = [1,2,3,4,5,6,7,8,9]
x = arr[5]
print(x)
```

### 2. O(log n) — Logarithmic Time
`The input is reduced by a factor each step (usually half).`

```python
x = 100
list = []

while x > 0:
    list.append(x)
    x //= 2

print(list)
```

### 3. O(√n) — Square Root Time
`Common when checking factors/divisors.`

```python
x = 10
list = []

for i in range(x):
    if i * i > x:
        break
    list.append(i)
    
print(list)
```

### 4. O(n) — Linear Time

`Work grows directly with input size.`

```python
n = 10
list = []

for i in range(n):
    list.append(i)
    
print(list)
```

### 5. O(n log n) — Linearithmic Time
`Usually seen in efficient sorting algorithms.`

```python
x = 10
list1 = []
list2 = []

for i in range(x):
    n = x
    list1.append(i)
    while n > 0:
        list2.over(n)
        n //= 2

print(list1)
print(list2)
```

### 6. O(n²) — Quadratic Time
`Usually caused by nested loops.`

```python
for i in range(10):
    for j in range(10):
        pass 
```

### 6. O(n²) — Quadratic Time
`Usually caused by nested loops.`

```python
for i in range(10):
    for j in range(10):
        for k in range(10):
            pass
```

### 8. O(2ⁿ) — Exponential Time
`Every step doubles the work.`
`Example: Naive Fibonacci`

```python
def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)

fib(6)
```

### 9. O(n!) — Factorial Time 🚨 Worst
`Common in brute-force permutation problems.`
