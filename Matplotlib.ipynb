## Pyplot API

`Univariate Analysis: Focuses only on one type of data`

```python
data = {
    'Salary': [10000,30000,20000,25000,21000,15000,20000,22000,22000,20000,45000,15000,10000,18000,30000,23000,28000,40000,37000,35000]
}
```

```python
df = pd.DataFrame(data)
df.shape
```

### `Line Plot`

```python
plt.plot(df['Salary'], color = 'red', marker = 'o', linestyle = '--', linewidth = '1') #linestyle = ':'
plt.grid()
#plt.axis('off')
plt.show()
```

### `Histogram`

```python
plt.hist(df['Salary'], bins= 7)
plt.show()
```

### `box Plot`

```python
plt.boxplot(df['Salary'])
plt.show
```

```python
df.loc[20] = 80000 #Created Outliers
plt.boxplot(df['Salary'])
plt.show
```

```python
df.drop(index= 20, inplace=True)
```

```python
df['Dept'] = ['IT', 'CS', 'AI', 'ML', 'DS', 'DS', 'ML', 'ML', 'AI', 'IT', 'CS', 'IOT', 'AI','AI','DS','ML','AI','DS','IT','CS']
df
```

```python
count = df['Dept'].value_counts()
count
```

```python
plt.pie(count, labels=count.index, autopct='%1.0f', explode=[0,0.1,0,0,0.5,0])
plt.axis('equal')
plt.show()
```

### `Count Plot`

```python
plt.bar(count.index, count)
plt.show()
```

```python
plt.bar(df['Dept'], df['Salary'], color = ['pink'])
plt.show()
```

`bivariate Analysis: Focuses only on two type of data`

```python
df['Age'] = [25,26,30,32,33,36,38,39,40,29,31,36,35,30,41,38,27,26,29,33]
df.tail()
```

### `Scatter Plot`

```python
plt.scatter(df['Age'], df['Salary']) #No Corelation
plt.show()
```

### `Line Plot`

```python
sort_age = df.sort_values('Age')
```

```python
plt.plot(sort_age['Age'], df['Salary'])
plt.grid()
plt.show()
```

### `Bar Chart`

```python
plt.bar(df['Age'], df['Salary'])
plt.grid()
plt.show()
```

`Bivariate Analysis: with Numerical and Categorical Data`

```python
cs = df[df['Dept'] == 'CS']['Salary']
it = df[df['Dept'] == 'IT']['Salary']
ai = df[df['Dept'] == 'AI']['Salary']
ml = df[df['Dept'] == 'ML']['Salary']
ds = df[df['Dept'] == 'DS']['Salary']
iot = df[df['Dept'] == 'IOT']['Salary']
```

```python
#Box Plot
plt.boxplot([ai,cs,ds,ml,it,iot], tick_labels=['ai','cs','ds','ml','it','iot'])
# plt.legend()
plt.show()
```

```python
# Pie chart
group = df.groupby('Dept')['Salary'].sum()

plt.pie(group, labels=group.index, autopct='%1.1f') #
plt.show()
```

```python
#Bar plot
cs_mean = sum(cs)/len(cs)
ai_mean = sum(ai)/len(ai)
ml_mean = sum(ml)/len(ml)
ds_mean = sum(ds)/len(ds)
it_mean = sum(it)/len(it)
iot_mean = sum(iot)/len(iot)

plt.bar(['cs','it','ai','ds','ml','iot'], [cs_mean, it_mean, ai_mean, ds_mean, ml_mean, iot_mean])
plt.show
```

`Multivariate Analysis: Focuses on multiple data`

```python
df['Exp']= [1,3,2,2.5,2.1,1.5,2,2.2,2.2,2,4.5,1.5,1,1.8,3,2.3,2.8,4,3.7,3.5]
df.head()
```

### `Bubble Chart`

```python
plt.scatter(df['Age'], df['Salary'], df['Exp']*30, color= 'skyblue', edgecolors='blue')
plt.title('Age vs Salary vs Experience')
plt.xlabel('Age')
plt.ylabel('Salary')
#plt.grid()
plt.show()
```

```python
plt.scatter(df['Salary'], df['Exp'], c = df['Dept'].map({'CS': 'red', 'AI': 'blue', 'IT': 'yellow', 'DS': 'black', 'ML': 'Green', 'IOT': 'grey' })) #color= 'skyblue', edgecolors='blue')
plt.title('Age vs Salary vs Experience')
plt.xlabel('Age')
plt.ylabel('Salary')
#plt.grid()
plt.show()
```

```python
color = {'CS': 'red', 'AI': 'blue', 'IT': 'yellow', 'DS': 'black', 'ML': 'Green', 'IOT': 'grey' }

for dept, color in color.items():
    df_dept = df[df["Dept"] == dept]
    plt.scatter(df_dept['Salary'], df_dept['Exp'], c = color, label = dept)

plt.legend()
plt.show()
```

## Object Oriented API

```python
fig, axis = plt.subplots(1,3, figsize =(20,5))


axis[0].plot(sort_age['Age'], sort_age['Salary'])
axis[0].set_title('Line Plot')

axis[1].scatter(sort_age['Age'], sort_age['Salary'])
axis[1].set_title('Scatter Plot')

axis[2].pie(group, labels=group.index, autopct='%1.1f')
axis[2].set_title('Pie Chart')

plt.savefig('Multiple_Plots.png')

plt.show()
```
